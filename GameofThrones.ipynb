```python
#library for data manipulation and analysis
import pandas as pd
#library for scientific computing
import numpy as np
#library for data visualisation
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#read csv file
battleofthrones = pd.read_csv('battles.csv')
```

```python
#see first few rows of dataset
battleofthrones.head()
```

```python
#see size of dataset
battleofthrones.shape
```

```python
#change name of column for attacker
battleofthrones.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
battleofthrones.head()
```

```python
#change name of column for defender
battleofthrones.rename(columns={'defender_1':'primary_defender'},inplace=True)
battleofthrones.head()
```

```python
#See number of occurences for attacker king
battleofthrones['attacker_king'].value_counts()
```

```python
#See number of occurences for locations
battleofthrones['location'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king',y='attacker_size',data=battleofthrones, palette='muted')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='defender_king',y='defender_size',data=battleofthrones, palette='muted')
plt.show()
```

```python
sns.countplot(x=battleofthrones['attacker_king'],hue=battleofthrones['battle_type'])
plt.show()
```
